```python
age_0_20 = ages <= 20
age_21_64 = (ages > 20) & (ages <= 64)
age_65_plus = ages > 64
```

```python
plt.figure(figsize=(10, 8))

plt.fill_between(ages[age_0_20], population[age_0_20], color='yellow')
plt.fill_between(ages[age_21_64], population[age_21_64], color='blue')
plt.fill_between(ages[age_65_plus], population[age_65_plus], color='deeppink')
plt.text(5, 10, '0 to 20 Years\n512 Mn\n36.1%', fontsize=10)
plt.text(30, 12, '21 to 64 Years\n807 Mn\n57.0%', color='white', fontsize=10)
plt.text(75, 4, '65+ Years\n98 Mn\n6.9%', fontsize=10)
plt.title("Indiaâ€™s Population distribution by Age in 2022", fontsize=14, fontweight='bold')
plt.suptitle("Population 1.42 billion\nMedian Age 28", fontsize=10, y=0.89)

plt.xticks(np.arange(0, 101, 5))
plt.yticks(np.arange(0, 31, 5), [f"{i} Mn" for i in range(0, 31, 5)])
plt.xlabel("Age")
plt.ylabel("Population")
plt.grid(False)
plt.box(False)
plt.savefig("India_Population_Distribution_Chart_2022.png", dpi=300, bbox_inches='tight')
plt.show()
```
